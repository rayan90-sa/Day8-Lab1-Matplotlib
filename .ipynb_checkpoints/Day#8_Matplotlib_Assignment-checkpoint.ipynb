```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Q1: Create two charts using list and dataframe

```python
# list 
plt.plot([10,20,30,40]);
# dataFrame
plt.plot(pd.DataFrame([100,200,300,400] ));
```

## Q2: Create pie chart with equal explode values.

```python
x=np.arange(1,5)
plt.pie(x , labels= ['A','B','C','D'] ,explode = [0.1 , 0.1 , 0.1, 0.1]);
# plt.title("Sales" , fontsize= 20 , loc= 'left' , color= 'm' );
```

## Q3: Create a scatter chart with circle marker, alpha = 0.4, and size = 50.

```python
x=np.random.randn(100)
y=np.random.randn(100)
plt.scatter(x,y , alpha=0.4 , s=50);
```

## Q4: Add the following to your chart:
- title placed in the left and has a red color
- x and y label with font size = 16

```python
x=np.arange(10)
y=np.arange(10)
plt.xlabel("Sales")
plt.ylabel("months")
plt.bar(x,y);
plt.title("Lab_8" , fontsize= 16 , loc= 'left' , color= 'r');
```

## Q5: Create three line charts with Legends at left and colors

```python
plt.plot([1,2,3,4]);
plt.plot([5,6,7,8]);
plt.plot([9,10,11,12])
plt.legend(['2009', '2008']);
plt.title( "Q5_Lab_8",loc= 'left' , color= 'y');
```

## Q6: Create subplots with three charts and figsize=20,10

```python
plt.subplot(2,2,1)#(row , col , postion of imges)
plt.plot([1,2,3,4]);
plt.plot([5,6,7,8]);
plt.legend(['2009', '2008']);

plt.subplot(2,1,2)
x=np.arange(1,5)
plt.pie(x , labels= ['A','B','C','D']);
plt.subplot(2,2,2)
x=np.arange(1,5)
plt.pie(x , labels= ['A','B','C','D']);

plt.title("Sales" , fontsize= 20 , loc= 'left' , color= 'm');
```
